```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import os

# Robust path detection
base_path = "DATASET/LA/LA/ASVspoof2019_LA_train/flac/LA_T_1000137.flac"
possible_paths = [
    base_path,              # If running from project root
    os.path.join("..", base_path) # If running from notebooks/ folder
]

audio_path = None
for p in possible_paths:
    if os.path.exists(p):
        audio_path = p
        break

if audio_path:
    print(f"Loading file: {audio_path}")
    audio, sr = librosa.load(audio_path)

    mfcc = librosa.feature.mfcc(y=audio, sr=sr)

    plt.figure(figsize=(10,4))
    librosa.display.specshow(mfcc, x_axis="time")
    plt.colorbar()
    plt.title(f"MFCC Features - {os.path.basename(audio_path)}")
    plt.show()
else:
    print("Could not find the audio file. Please ensure you are running the notebook from the 'notebooks' folder or the project root.")
    print(f"Tried: {possible_paths}")
```
